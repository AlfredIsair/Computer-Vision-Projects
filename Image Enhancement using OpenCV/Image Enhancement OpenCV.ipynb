```python
#import necessary libraries
import cv2 
import matplotlib.pyplot as plt 
import numpy as np 
  
```

```python
#load the image
image =cv2.imread(".\Downloads\darth vader.jpg")
```

```python
#plot the original image
plt.subplot (1, 1, 1) 
plt.title("original")
plt.imshow(image)
```

### **Brighness and Contrast**

Adjusting the brigtness and contrast affects an image visual appeal and effectiveness.Finding the right balance of brightness and contrast is important for creating an attractive and effective image.

The **cv2.addWeighted()** functionis used to adjust the brighness and contrast

```python
#Adjusting brightness and contrast
brightness = 10
contrast = 2.3
image2 =cv2.addWeighted(image,contrast, 
                       np.zeros(image.shape,image.dtype),
                      0,brightness)
```

```python
#plot  the contrast image
plt.subplot(1, 1, 1)
plt.title("Brightness and contrast")
plt.imshow(image2)
```

Another method for adjusting the brightness and contrast of an image is to use the *cv2.convertScaleAbs()* function, which allows  to adjust the brightness and contrast using a combination of scaling and shifting the pixel values.

```python
#Adjusing the brightness and contrast using the formula g(i,j)=f(i,j)+ 
alpha = 1.5 #alpha is the contrast
beta  = 40  #beta is the brightness
image3 =cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
```

```python
#plot the contrast image
plt.subplot(1, 1, 1)
plt.title("Brightness and contrast 2")
plt.imshow(image3)
```

### Image Sharpening
Involves enhancing edges and fine details in an image.Used to correct blur or softness in an image

**cv2.filter2D()** function is used which convolves the image with a kernel (small matrix that acts as a filter)

```python
#create the sharpening kernel 
kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])

#sharpen the image
sharpened_image= cv2.filter2D(image, -1, kernel)
```

```python
#Plot the original image 
plt.subplot(1, 2, 1) 
plt.title("Original") 
plt.imshow(image) 
  
#plot the sharpened image
plt.subplot(1, 2, 2)
plt.title("sharpened")
plt.imshow(sharpened_image)
plt.show
```

Another method for sharpening images is to use the cv2.Laplacian() function, which calculates the Laplacian of an image and returns the result as a sharpened image.

```python
#Plot the original image 
plt.subplot(1, 2, 1) 
plt.title("Original") 
plt.imshow(image) 
  
# Sharpen the image using the Laplacian operator 
sharpened_image2 = cv2.Laplacian(image, cv2.CV_64F) 

#Plot the sharpened image 
plt.subplot(1, 2, 2) 
plt.title("Laplacian Sharpening") 
plt.imshow(sharpened_image2) 
plt.show()
```

### Removing Noise from Image

Process of removing or reducing unwanted noise and artifacts from an image.
Methods used involve applying filters like Median filter and Gaussian filter to avoid blur or loss of detail in the image

**cv2.medianBlur()** function applies the Median filter which works by replacing each pixel in the image with the median value of a sset of neighbouring pixels

```python
# Remove noise using a median filter 
filtered_image = cv2.medianBlur(image, 5) 


#Plot the original image 
plt.subplot(1, 2, 1) 
plt.title("Original") 
plt.imshow(image)

#Plot the blured image 
plt.subplot(1, 2, 2) 
plt.title("Median Blur") 
plt.imshow(filtered_image)
plt.show
```

**cv2.GaussianBlur()** applies a Gaussian filter to the image which uses a weighted average of neighboring pixels to smooth out noise and reduce artifacts.

```python
#Remove the noise using Gaussian filter
#(9, 9) parameter specifies the size of the kernel to use for the filter, 
# 0 parameter specifies the standard deviation of the Gaussian function.
filtered_image2 = cv2.GaussianBlur(image, (9, 9), 0)

#Plot the original image 
plt.subplot(1, 2, 1) 
plt.title("Original") 
plt.imshow(image)

#Plot the blured image 
plt.subplot(1, 2, 2) 
plt.title("Gaussian Blur") 
plt.imshow(filtered_image2)
plt.show
```

### Enhancing Image Color

Adjusting color in an image makes them more vibrant , bold or natural.Techniques used include adjusting color balance, saturation and hue.

**cv2.cvtColor()** function allows to convert image from one color to another

```python
#Plot the original image 
plt.subplot(1, 2, 1) 
plt.title("Original") 
plt.imshow(image) 

# Convert the image from RGB to HSV color space 
image1 = cv2.cvtColor(image, cv2.COLOR_RGB2HSV) 

# Adjust the hue, saturation, and value of the image 
# Adjusts the hue by multiplying it by 0.7 
image1[:, :, 0] = image1[:, :, 0] * 0.7
# Adjusts the saturation by multiplying it by 1.5 
image1[:, :, 1] = image1[:, :, 1] * 1.5
# Adjusts the value by multiplying it by 0.5 
image1[:, :, 2] = image1[:, :, 2] * 0.5

# Convert the image back to RGB color space 
image2 = cv2.cvtColor(image1, cv2.COLOR_HSV2BGR) 

#Plot the enhanced image 
plt.subplot(1, 2, 2) 
plt.title("enhanced coloured") 
plt.imshow(image2) 
plt.show()
```

### Image Resizing and Scaling 
This involves adjusting the dimension and size of an image to fit a specific size or aspect ratio.
**cv2.resize()** function is used to state the size of the new image

```python
#Plot the original image 
plt.subplot(1, 2, 1) 
plt.title("Original") 
plt.imshow(image) 

# Resize the image to a specific width and height 
resized_image = cv2.resize(image, (5500, 4000)) 

#Plot the Resized image 
plt.subplot(1, 2, 2) 
plt.title("Resized") 
plt.imshow(resized_image) 
plt.show() 
```

use the fx and fy parameters to specify the scaling factors along the x-axis and y-axis

```python
#Plot the original image 
plt.subplot(1, 2, 1) 
plt.title("Original") 
plt.imshow(image) 

# Scale the image by a factor of 2 along both axes 
scaled_image = cv2.resize(image, None, fx=2, fy=2) 

#Plot the Scaled image 
plt.subplot(1, 2, 2) 
plt.title("Scaled") 
plt.imshow(scaled_image) 
plt.show()
```

### Inverse Transform
 inverse the color by simply subtracting each value from 255

```python
#Plot the original image 
plt.subplot(1, 2, 1) 
plt.title("Original") 
plt.imshow(image) 

# Inverse by subtracting from 255 
inverse_image = 255 - image 

#Save the image 
cv2.imwrite('inverse_image.jpg', inverse_image) 
#Plot the Inverse image 
plt.subplot(1, 2, 2) 
plt.title("Inverse color") 
plt.imshow(inverse_image) 
plt.show()
```

### Histogram Equalization
Technique used to adjust the contrast of an image by spreading the intensity values of image pixels.

Global histogram equalization adjuss contrast of the entire image while local histogram equalization adjusts contrast in small localized areas in an image.

**cv2.equalizeHist()** function takes the image as as rgument and returns the equalized image as the data

```python
#Plot the original image 
plt.subplot(1, 2, 1) 
plt.title("Original") 
plt.imshow(image) 

# Convert the image to grayscale 
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 

# Equalize the histogram 
equalized_image = cv2.equalizeHist(gray_image) 

#Save the equalized image 
cv2.imwrite('equalized.jpg', equalized_image) 

#Plot the equalized image 
plt.subplot(1, 2, 2) 
plt.title("equalized") 
plt.imshow(equalized_image) 
plt.show()
```
